### 1. Rewrite your pay computation to give employee 1.5 times the hourly rate if he has worked for more than 40 hours.
#### _Let following be the input and output be following constraints_
#### 1.Take input for no. of hours worked
#### 2. Take input for hourly rate
#### 3. Output the total amount to be paid to employee

```python
#function to count no of arguments of a function
#This function is not much used in the code below
def count_args(*args):
    i = 0
    while(1):
        try :
            temp = args[i]
            if temp != None:
                i += 1
                print(temp)
        except:
            break
    return i


'''
this function takes func as an input 
which is a function which gets input from 
the user through CLI and returns list of 
those values. 
check_int function checks if those values 
are convertible to integer format and if they 
are convertible it converts those values 
into integer format 

'''
def check_int(func):
    def inner(*args, **kwargs):
        val = func(*args, **kwargs)
        flag = 0
        i = 0
        try:
            while val[i] != None :
                try:
                    val[i] = int(val[i])
                except:
                    print("Give integer as an input")
                    flag = 1
                    break
                i += 1
            if flag == 1:
                return inner(*args, **kwargs)
        except :
            return val
    return inner      
        

@check_int
def input_val():
    hours = input("Enter the no of hours: ")
    rate = input("Enter the hourly rate: ")
    return [hours, rate]


def emp_salary():
    [hour, rate] = input_val()
    if hour > 40:
        rate *= 1.5
    amount = hour*rate #amount to be paid to the employee
    return amount

if __name__ == "__main__":
    salary = emp_salary()
    print(f"salary :{salary}")
```

### Example :-
**In the last example of the dictionary that we saw we knew the keys of the dictionary from which we calculated the total amount in the purse**
    **Let us assume we have a long json data of notes in purse which we converted to the dictionary.**
    **Let's write a function to calculate the total amount in a purse or wallet.**
    **Here we don't know the keys of the dictionary.** **We can make use of** *keys()* **function to get the keys in the dictionary.**

```python

def total_money(wallet):
    total = 0
    keys = wallet.keys()
    for i in keys:
        try:
            total += i*wallet[i]
        except TypeError:
            try :
                key = int(i)
                try :
                    value = int(wallet[i])
                except ValueError:
                    print("Give a valid input dictionary containing numerical literal values and keys")
                    print("default key: ",i)
                    return None
                    
            except ValueError:
                print("Give a valid input dictionary containing numerical literal values and keys")
                print("default key: ",i)
                return None
            total += key*value
        
    return total
purse = {10:5,20:"jay",30:2}
a = "jay"
print(total_money(purse))

```

### write a counter to count the maximum number of repitition of a string in a list
`a = ["jayesh", "jay", "jay", "sam","soham","soham","sam","jayesh","jay","sam","soham"]`


```python
def str_counter(x):
    b = dict()
    for i in x:
        b[i] = b.get(i, 0) + 1
    return b
a = ["jayesh", "jay", "jay", "sam","soham","soham","sam","jayesh","jay","sam","soham"]
print(str_counter(a))
```

```python
fname = "C:/Users/jayes/source/repos/pythonDS/eg_test.txt"

def file_word_count(filename):
    fhand = open(filename, 'r')
    word_count = dict()
    for line in fhand:
        words = line.split()
        for word in words:
            word_count[word] = word_count.get(word, 0) + 1
    fhand.close()
    word_count = filter_num(word_count)
    return word_count

#function to determine the maximum count of the string 
def max_count(a):
    maxcount = None
    max_count_name = None
    for key, value in a.items():
        if maxcount is None or maxcount < value :
            maxcount = value
            max_count_name = key
    return [max_count_name, maxcount]

#function for determining if a string is integer convertible
def check_int(a):      # a is of type string
    try:
        int(a)
        return True
    except TypeError:
        return False
    except ValueError:
        return False
        
    
#function for filtering the number from the dictionary of the counts and only taking string having alphabets and symbol
def filter_num(x): #x is of type dict
    del_keys = list()
    for key in x:
        if '.' in key:
            temp = key.split('.')
            for i in temp:
                flag = check_int(i)
                if flag == True:
                    del_keys.append(key)
                    break;
                
                    
        elif ':' in key:
            temp = key.split(':')
            for i in temp:
                flag = check_int(i)
                if flag == True:
                    del_keys.append(key)
                    break;
        else:
            flag = check_int(key)
            if flag == True:
                del_keys.append(key)
    
    for del_key in del_keys:
        del x[del_key]
    return x

if __name__ == "__main__":
    #here we will be doing these steps
    '''
    1. we will read the file and find the word count for each number
    2. while finding the word count we make sure that we do not include the integers in our dictionary
    3. then we output the word count
    4. we find the word with maximum word count and print it to the console
    '''
    result = file_word_count(fname)
    max_repitition = max_count(result)
    print(result)
    print(max_repitition)
```
